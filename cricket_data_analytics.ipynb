we can observe that there is details of 3891 match among which 50% match occured before 2003 inorder to analyse india's performance of this year. that old data will be irrelavant. therefore decrease the data to have deeper understanding  

```python
matches_007 = matches[matches.match_year>=2007]
matches_007.head()
```

```python
matches_007.count()
```

```python
matches_007.Team_1.unique() #got the list of teams..to know india performnce we need to decrease the lines in plot.
```

```python
Top_teams = ['Australia', 'England', 'India', 'bangladesh', 'New Zealand', 'Pakistan', 'South Africa', 'Sri Lanka', 'West Indies' ]
#decreased no. of teams
matches_007 = matches_007[matches_007.Team_1.isin(Top_teams)] #selelct entries in team1 column that have matching teams as of top_teams
matches_007 = matches_007[matches_007.Team_2.isin(Top_teams)]
matches_007.count()
```

```python
matches_007.head()
```

take eg. of India, india can be in team1 and team2 so it is neccesary to take team1 winner and team2 winner then merge both

```python
matches_007['team_1_win'] = matches_007.Team_1 == matches_007.Winner #list winner from each column and create new columnn called team_1_win
matches_007['team_2_win'] = matches_007.Team_2 == matches_007.Winner
matches_007.head()
```

```python
#aggregating the data with total no. of match and total win by each team
matches_007.groupby('Team_1').agg({'Scorecard': 'count', 'team_1_win': 'sum'})
```

```python
#now we get team wise to that year matches played(scorecard) and match won(team_1) similarly for team2
team_1 = matches_007.groupby(['Team_1', 'match_year']).agg({'Scorecard': 'count', 'team_1_win': 'sum'}).reset_index()
team_2 = matches_007.groupby(['Team_2', 'match_year']).agg({'Scorecard': 'count', 'team_2_win': 'sum'}).reset_index()
team_1.head()

#reset_index() method to reset a multi-level index. For example, if you have a dataframe with a multi-level index on the columns, you can use the reset_index() method to reset the index to a single level.
```

```python
#resetindex gives index 0123
team_2.head()
```

```python
team_1 = team_1.rename(columns = {'Team_1':'team', 'Scorecard': 'total_matches'})
team_2 = team_2.rename(columns = {'Team_2':'team', 'Scorecard': 'total_matches'})
```

```python
team_1.head()#changed name of score card to total_matches
```

```python
team_2.head()
```

MERGE = #merge bcz india won few matches when it was part of team_1 and won some match as part of tem_2 and it done on the year and team1 and 2 to team


```python
team_info = pd.merge(team_1, team_2, on = ['team', 'match_year'],how='outer')
print(team_info.count())
team_info.head()
```

```python
team_info = team_info.fillna(0) #replace all null values with specified values
team_info.head()
```

```python
team_info['total_matches'] = team_info['total_matches_x']+ team_info['total_matches_y']
team_info['total_wins'] = team_info['team_1_win']+team_info['team_2_win']
team_info.tail(30)
```

```python
team_info['win_perc'] = 100* team_info['total_wins']/team_info['total_matches']
team_info.head()

#creating a column telling winning percentage
```

```python
#create filter to reduce lines in graph
team_info_x = team_info[team_info.team.isin(['India', 'Pakistan', 'Australia', 'England'])]
```

```python
team_info_x.pivot(index= 'match_year', columns = 'team', values = 'win_perc').plot.line()
#creating line graph for teams using their winning perc
```
